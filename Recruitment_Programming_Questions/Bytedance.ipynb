## 1. Magic

1. n个正整数集合a, m个正整数的集合b
2. 从一个集合移动一个元素到另一个集合中，保证两个集合的平均值变大！
3. 求最大移动次数
4. 实例
        输入：
        3 5
        1 2 5
        2 3 4 5 6
        输出：
        2
        
### 分析：
平均值的求法 ：
定义s(n) 为集合中前n个元素的和，则 s(n) == (s(n-1) * (n-1) + a_n ) / n

从平均值大的集合a中去比平均值小且并b平均值大的数放入b中

```python
from heapq import heappop, heappush, heapify

def find_min(mylist, big_mean, small_mean):
    '''
    @param mylist : sorted list
    @big_mean : int
    @small_mean : int
    '''
    ret, i = -1, 0
    while mylist:
        cur = heappop(mylist)
        if cur < big_mean and cur > small_mean:
            ret = cur
            break
    return ret

def magic_move_num(list_a, list_b):
    m, n = len(list_a), len(list_b)
    sum_a, sum_b = sum(list_a), sum(list_b)
    mean_a = sum_a / m
    mean_b = sum_b / n
    #the average of set a is bigger than that of b
    if mean_a < mean_b:
        list_a, list_b = list_b, list_a
        m, n = n, m
        mean_a, mean_b = mean_b, mean_a
        sum_a, sum_b = sum_b, sum_a
    heapify(list_a)
    #calculate the step
    step = 0
    min_num = find_min(list_a, mean_a, mean_b)
    while min_num != -1:
        sum_a -= min_num
        sum_b += min_num
        m, n = m-1, n+1
        mean_a = sum_a / m
        mean_b = sum_b / n

        step += 1
        
        min_num = find_min(list_a, mean_a, mean_b)
    return step


list_a = [1, 2, 5]
list_b = [2, 3, 4, 5, 6]
magic_move_num(list_a, list_b)
```

## 3. 矩形游戏

1. 规定 u,d,l,r 代表向上一个，向下，向左，向右4种操作
2. 给定一个N * M 的空地，起点(x,y), 以及一系列操作集L，如果走出矩形区域则停止
3. 输出每种 (N,M,x,y)场景下，按照L操作后可行的步数（包含最后一次）
4. 实例
        输入
        uuurrdddddl
        3
        5 6 3 3
        5 6 4 2
        6 6 4 2
        输出
        3
        10
        11
        
### 分析

1. 问题不难，可以直接暴力求解，如果想要更快的速度，优化存储和判断的逻辑
2. 用四维数组max_size表示U,D,L,R四个方向，每个操作能到达当前方向的最大范围 l[i+1] = max(l[i], -y) r[i+1] = max(r[i], y)
3. 用四维数组max_step表示U,D,L,R四个方向, 每增加一个单位能完成的操作数
4. 四个方向中的最小值就是最大移动步数

```python
def matrix_game(ops_str, maze_list):
    # U D L R
    ops_len = len(ops_str)
    max_size = [[0] * (ops_len + 1) for _ in range(4)]
    x, y = 0, 0
    for i, ops in enumerate(ops_str):
        if ops == 'u': x -= 1
        elif ops == 'd' : x += 1
        elif ops == 'l' : y -= 1
        elif ops == 'r' : y += 1
        for direction, step in enumerate([-x, x, -y, y]):
            max_size[direction][i+1] = max(max_size[direction][i], step)
    for i in range(4):
        print(max_size[i])
        
    max_step = [[0] * (ops_len+1) for _ in range(4)]
    for direction in range(4):
        for i in range(1, ops_len):
            j = 0
            while j < ops_len and max_size[direction][j] < i :
                j += 1
            max_step[direction][i] = j
    for i in range(4):
        print(max_step[i])
        
    for maze in maze_list:
        maze = list(map(int, maze.split(' ')))
        n, m, x, y = maze
        max_u = ops_len if x > ops_len else max_step[0][x]
        max_d = ops_len if n - x + 1 > ops_len else max_step[1][n - x + 1]
        max_l = ops_len if y > ops_len else max_step[2][y]
        max_r = ops_len if m - y + 1 > ops_len else max_step[3][m-y+1]
        print(min(max_u, max_d, max_l, max_r))
ops = 'uuurrdddddl'
maze_list=[
    '5 6 3 3',
    '5 6 4 2',
    '6 6 4 2'
]
matrix_game(ops, maze_list)
```

### 4.1 扩展欧几里得算法
在计算出gcb(a,b) 的同时，找到a * x + b * y = gcb (a, b) 最小整数解(x,y)
证明：

b = 0, gcb(a, b) = a, 此时 x=1, y=0

设：
$$
\begin {aligned}
&a * x_1 + b * y_1 = gcb(a, b) \\
&b * x_2 + b * (a \bmod b) = gcb(b, a \bmod b) 
\end {aligned}
$$
因为 $ gcb(a, b) == gcb(b, a \bmod b) $

所以
$$
\begin {aligned}
a * x_1 + b *  y_1 &= b * x_2 + (a \bmod b) * y_2 \\
&=b * x2 + (a - (a //b) * b) * y_2 \\
&=a * y_2 + b * x_2 - (a // b) * b * y_2
\end {aligned}
$$
根据待定系数法
$$
\begin {aligned}
&x_1 = y_2 \\
&y_1 = x_2 - (a // b) * y_2
\end {aligned}
$$

```python
def exgcb(a, b):
    if b == 0:
        #递归基, b=0, gcb(a,b)=a, x=1, y=0
        return 1,0,a
    #进入递归
    x,y,q = exgcb(b, a % b)
    #old_x
    tmp = x
    #print(x, y)
    #new x and new y
    x = y
    y = tmp - a // b * y
    return x, y, q

exgcb(5,8)
```

### 4.2 一种可行的方法
1. 给A充电;
2. A转移到B;
3. 如果B满电，则放空电量;
4. 如果A非空，则重复2-4;
5. 从1开始重复整个过程，直到A或B中的电量是 c为止。

```python
#https://www.acwing.com/blog/content/17/
#https://blog.csdn.net/jiaobuchong/article/details/40402885
#一种可行的方法
def test(ca, cb, target):
    step = 0
    a, b = 0, 0
    while a != target and b != target:
        print("a : %d, b = %d"%(a, b))
        if a == 0:
            a = ca
            print("charge the container a with %d"%a)
        
        if a + b <= cb:
            b += a
            a = 0
        else:
            a -= (cb - b)
            b= cb
        print("after transfer from a to b : a=%d b=%d"%(a,b))
        if a == target or b == target:
            break
        if b == cb:
            print("reset b(%d)"%b)
            b = 0
            
        step += 1
        if step > 100:
            break
test(8,5,4)
```

首先如果 c>a 并且 c>b 时，操作显然无法完成。所以我们只考虑 c≤Max(a,b)的情况。

我们将两个电容看做一个整体，与外界只有两种操作：

    充电和放电。充电会使整个系统的总电量增加，放电会使整个系统的总电量减少。

我们先判断整个系统的总电量是否可以等于 c，即是否存在整数 x,y，使得 ax+by=c。

如果 x,y 是负整数，则表示放电，如果是正整数，则表示充电。根据裴蜀定理，等式可以满足当且仅当 gcd(a,b)|c

我们现在来考虑当 gcd(a,b)|c时，如何计算最少操作次数：

    根据裴蜀定理，存在整数 x,y，使得 ax+by=c，且由于 c≤Max(a,b)，所以 x,y 一定一正一负，或者一正一零。 

根据4.2流程：

        给A充电;
        A转移到B;
        如果B满电，则放空电量;
        如果A非空，则重复2-4;
        从1开始重复整个过程，直到A或B中的电量是 c为止。
        
如果 c>a，则B里先有 c 的电量（因为A存不下），此时有 x次充电，−y次放电（y是负数），x−y次转移，共 2x−2y 次操作；

如果 c≤a，假设B中先有 c 的电量，则 a,b≥c，所以 b=c，仅需一次操作；

假设A中先有 c 的电量，则有 x 次充电，x−y−1 次转移，−y−1次放电，共 2x−2y−2次操作；

因此我们只需最小化 |x|+|y|。可以先用扩展欧几里得算法求出任意一组 (x,y)，则
$$
\left\{ 
\begin{aligned} 
x + kb/gcd(a,b) \\ 
y - ka/gcd(a,b) 
\end{aligned} 
\right. 
, k = …, -2, -1, 0, 1, 2, …
$$
就是该等式的全部整数解，我们找出和 00 最接近的两组 (x,y)（一组 x>0, 另一组 y>0 ），然后取操作次数较小的一组即可。

时间复杂度分析：扩展欧几里得算法的复杂度是 O(logV)，一共 N 组测试数据，所以总时间复杂度是 O(NlogV)。



```python
def least_ops(a, b, c):
    x,y,d = exgcb(a,b)
    '''
    考虑 y < 0 时的情况，既a充，b放
    '''
    if y>0:
        x,y = y,x
        a,b = b,a
    #将a,b等比例缩小直到两数互质
    a2, b2 = a/d, b/d
    #得到一个组a * x + b * y = c 的解(x,y)
    x *= c / d
    y *= c / d
    #根据 a(x-kb) + b(y-ka) == ax + by, 缩小 x,y
    k = x // b2
    x -= k * b2
    y += k * a2

    if c > a:
        #b电容先达到c值
        #a充电x，b放电-y
        #a中所有电都给y，此外遇到b被充满的情况，a需要再传一次才能将a转空
        #由于a,b互质，所以刚好需要-y 次，所以传递需要 x-y次是，共计x-y+(x-y)
        res = 2 * (x-y)
    else:

        #a电容先到达c值
        #与上一情况相比，a充电x次, 转移次数少一（最后停到了a），重置次数少1（不太懂！）

        res = 2 * (x-y-1)
    '''
    考虑另一个解得情况，此时y > 0, x < 0，a放，b充
    '''
    x -= b2
    y += a2
    if c > b:
        res = min(res, 2*(y-x))
    else:
        res = min(res, 2 *(y-x-1))
    return res

least_ops(5, 8,4)
```
