```python
class DynamicArray():
    def __init__(self):
        self.min = MinHeap()
        self.max = MaxHeap()
    def insert(self, num):
        if (len(self.min.data) + len(self.max.data)) & 1 == 0:
            if self.max.size() > 0 and num < self.max.top():
                self.max.insert(num)
                num = self.max.pop()
            self.min.insert(num)
        else:
            if self.min.size() > 0 and self.min.top() < num:
                self.min.insert(num)
                num = self.min.pop()
            self.max.insert(num)
        #print('max = ', self.max.data)
        #print('min = ', self.min.data)
    def getMedian(self):
        size = self.min.size() + self.max.size()
        assert size > 0, "No numbers are available"
        median = 0
        if size & 1 == 1:
            median = self.min.top()
        else:
            median = (self.min.top() + self.max.top()) / 2.0
        return median

array = DynamicArray()
length = 100
data = list(range(length))
org = data.copy()
def getMedain(array):
    median = 0
    if length & 1 == 1:
        median = array[length//2]
    else:
        median = (array[length//2] + array[length//2 - 1]) / 2.0
    return median

#data = [33, 53, 82, 17, 57, 14, 31, 19, 35, 61, 96, 93, 49, 52, 95, 1, 16, 7, 2, 75, 88, 67, 5, 0, 41, 23, 90, 3, 58, 83, 73, 85, 71, 60, 28, 74, 51, 4, 98, 15, 62, 76, 34, 29, 40, 18, 56, 39, 24, 36, 81, 68, 89, 37, 13, 92, 64, 79, 55, 54, 9, 27, 12, 59, 21, 65, 6, 8, 97, 80, 22, 48, 50, 32, 70, 63, 78, 69, 91, 10, 11, 44, 43, 77, 30, 86, 20, 38, 25, 42, 72, 87, 47, 26, 99, 66, 46, 84, 94, 45]
for i in range(1000):
    random.shuffle(data)
    #print(data)
    print('\r%05d'%i, end='')
    for d in data:
        array.insert(d)
    #print(heap.data)
    m = array.getMedian()
    test = m == getMedain(org)
    if test == False:
        print(m, getMedain(org))
        print(data)
    #print(m, getMedain(org))

if test:
    print("\ntest passed")
```
