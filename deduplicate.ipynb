```python
import os
import re
import sys
```

```python
root= r'F:\资料整理\201836000L290121王豆豆'
path = r'F:\资料整理\201836000L290121王豆豆\title-demestic.txt'
path2 = r'F:\资料整理\201836000L290121王豆豆\title-foreign.txt'
os.path.exists(path2)
```

```python
pattern = r'\[\d+\]'
pattern2 = r'\d+\. \w+'
def load_file(pattern, path):
    title = []
    with open(path, 'r', encoding='utf-8')  as f:
        for row in f.readlines():
            if re.search(pattern, row):
                tmp = row.split('. ')
                if len(tmp) == 2:
                    row = tmp[1]
                elif row.find(']') != -1:
                    row = row[row.find(']')+1:]
                #row = row.replace('-',' ')
                title.append(row)
                
    return sorted(title)
```

```python
def deduplicate(title_list):
    i, j, length = 0, 1, len(title_list)
    while j < length:

        if title_list[i] != title_list[j]:
            i += 1
            title_list[i] = title_list[j]
                
        j += 1
    return title_list[:i+1]

title_demestic = load_file(pattern, path)
title_demestic_clear = deduplicate(title_demestic)
title_foreign = load_file(pattern2, path2)
title_foregin_clear = deduplicate(title_foreign)
print(len(title_demestic))
print(len(title_demestic_clear))
print(len(title_foreign))
print(len(title_foregin_clear))
```
